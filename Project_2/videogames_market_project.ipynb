## Предобработка данных

### Приводим к нижнему регистру названия столбцов

```python
data.columns = data.columns.str.lower() # используем соответствующую команду
```

### Преобразуем данные в нужные типы

```python
data.info() # выводим общую информацию о датафрейме
```

Данные в столбце year_of_release должны быть формата int64 (целое число), в столбцe user_score - float64.

```python
data['user_score'] = pd.to_numeric(data['user_score'], errors='coerce') # меняем формат данных на float64
```

Неподходящие значения (tbd) целесообразно при изменении формата данных заменить на пропуски, чтобы в дальнейшем столбец с оценками пользователей было проще обрабатывать (например, строить графики). По сути значение tbd также как и пропуск никакой смысловой нагрузки не несет.

Поменять формать данных в столбце year_of_release на int сразу мы не можем, так как там имеются пропуски. Целесообразно удалить строки с пропусками в столбце с годом (принимая во внимание их несущественное количество - менее 2%). Подобрать логичную замену пропускам в данной ситуации невозможно. Причина таких пропусков - техническая ошибка при экспорте данных.

```python
data = data.dropna(subset=['year_of_release']) # удалим строки с пропусками в столбце year_of_release, учитывая их несущественное количество
```

```python
data['year_of_release'] = data['year_of_release'].astype(int) # меняем формат данных на int64
```

### Обрабатываем пропуски

```python
data.isna().sum() # снова определяем, в каких столбцах есть пропуски, а также их количество
```

Мы видим, что у **несущественного** для нашей выборки количества игр есть пропуски в столбцах name (2) и genre (2). Такие строки **целесообразно удалить**, принимая во внимание, что подобрать логичную замену пропускам в данной ситуации невозможно. Причина таких пропусков - техническая ошибка при экспорте данных. При этом пропусков в столбцах critic_score, user_score и rating **относительно много** - больше 40%, поэтому удалять соответствующие строки **нельзя** (это существенно исказит результаты анализа).

```python
data = data.dropna(subset=['name']) # удалим строки с пропусками в столбце name 
```

```python
data = data.dropna(subset=['genre']) # удалим строки с пропусками в столбце genre
```

```python
no_rating = data[data['rating'].isnull()] # отфильтруем игры с пропущенным значением рейтинга
```

```python
no_rating.isna().sum() # определяем, в каких столбцах отфильтрованной таблицы есть пропуски, а также их количество 
```

```python
data['rating'].value_counts() # посмотрим, какие в целом значения имеются в столбце с рейтингом
```

Мы видим, что большая часть пропусков **пересекается** - в более чем 6 500 строк отсутствуют данные и по рейтингу, и по оценке критиков, и по оценке пользователей. Возможно предположить, что отсутствие рейтинга у небольшого ряда игр обусловлено тем, что они были выпущены до 1994 года (год основания ESRB - **североамериканской** ассоциации, определяющей возрастной рейтинг компьютерных игр).
Отсутствие рейтинга ESRB у остальных игр может быть обусловлено тем, что рейтинг им был присвоен иными ассоциациями (например, европейской PEGI или японской CERO).
Учитывая большое количество пропусков и невозможность их адекватной замены, целесообразно проставить вместо пропусков в столбце с рейтингом значение "No data" (так как остальные значения в столбце - также текстовые).

```python
data['rating'] = data['rating'].fillna('No data') # заполняем пропуски в столбце rating
```

### Исследуем влияние на продажи отзывов пользователей и критиков

```python
ps4_2 = data_new.query('platform == "PS4"') # сделаем срез данных по самой популярной в 2013-2016 гг. платформе - PS4
```

```python
ps4_2.plot(x='user_score', y='total_sales', xlabel='Оценка пользователей', ylabel='Общие продажи, млн копий', title='Распределение игр для PS4 по общим продажам и оценке пользователей', style='o', kind='scatter', color='green', xlim=(0, 10), ylim=(0, 10), grid=True, figsize=(8, 10)) # построим диаграмму рассеяния
```

Судя по диаграмме, связи между оценкой пользователей и продажами нет - мы видим, что есть большое количество коммерчески успешных игр, средняя оценка которых меньше 6 из 10. Например, самая продаваемая игра в выборке (чуть меньше 9 млн копий) вообще имеет оценку чуть больше 4 из 10. Одновременно мы видим довольно существенное количество высокооцененных игр (больше 8 из 10), продажи которых не перешагнули отметки в 1 млн копий.

```python
print(ps4_2['total_sales'].corr(ps4_2['user_score'])) # определим коэффициент Пирсона
```

Корреляция (коэффициент Пирсона) практически отсутствует (-0.03) - таким образом, либо довольные игроки в целом реже ставят оценки, чем недовольные, либо на привлекательность игры для потребителей оценки других игроков особого влияния не оказывают (учитывая, что данные оценки появляются через некоторое время после старта продаж и вообще не влияют на предзаказы). К тому же следует помнить, что пользователи ставят оценки играм уже после их релиза. Кроме того, средняя оценка пользователей может быть предметом манипуляций, например, со стороны "хейтеров" и фанатов конкурирующих франшиз.

```python
ps4_2.plot(x='critic_score', y='total_sales', xlabel='Оценка критиков', ylabel='Общие продажи, млн копий', title='Распределение игр для PS4 по общим продажам и оценке критиков', color='green', style='o', kind='scatter', xlim=(0, 100), ylim=(0, 10), grid=True, figsize=(12, 8)) # построим следующую диаграмму рассеяния
```

```python
print(ps4_2['total_sales'].corr(ps4_2['critic_score'])) # определим коэффициент Пирсона
```

Корреляция между продажами и оценкой критиков значительно сильнее, чем между продажами и оценкой пользователей, однако все равно не слишком велика (0.4). На графике видно, что игры с продажами более 4 млн копий не имеют оценку критиков ниже 70 (почти все эти игры получили среднюю оценку более 80). Вместе с тем, большое количество высокооцененных игр (оценка выше 70-80) не было коммерчески успешным (продажи менее 1 млн копий) и продались тиражом примерно таким же, как и игры условно худшего качества (оценка критиков ниже 70). Любопытно, что самая продаваемая игра (около 9 млн копий) не может похвастаться рекордно высокой оценкой критиков (около 82 из 100).

Вместе с тем, необходимо помнить, что количество пропусков в столбцах с оценками пользователей и критиков весьма существенное - если бы на их месте были реальные оценки, выводы могли бы быть другими.

## Проверим ряд гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

```python
data_new_hypothesis = data_new.dropna(subset=['user_score']) # сначала оставим в датафрейме только строки без пропусков в столбце с оценкой пользователей
```

```python
xone_hyp = data_new_hypothesis.query('platform == "XOne"') # сделаем срез по платформе
```

```python
pc_hyp = data_new_hypothesis.query('platform == "PC"') # сделаем второй срез по платформе
```

В данной ситуации мы имеем дело с гипотезой о равенстве средних двух генеральных совокупностей. Применим критерий scipy.stats.ttest_ind(), так как генеральные совокупности не зависят друг от друга, а выборочные средние нормально распределены.
Сформулируем нулевую гипотезу - средние пользовательские рейтинги игр для платформ Xbox One и PC одинаковые.
Сформулируем альтернативную гипотезу - средние пользовательские рейтинги игр для платформ Xbox One и PC разные.

```python
from scipy import stats as st
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
results = st.ttest_ind(
    xone_hyp['user_score'], 
    pc_hyp['user_score'], equal_var = False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Мы использовали дополнительный параметр equal_var = False, потому что у нас недостаточно большие и неодинаковые по размеру выборки - есть основания полагать, что разброс будет сильно отличаться.
Таким образом, полученные данные **не противоречат** гипотезе о том, что средние пользовательские рейтинги игр для платформ Xbox One и PC **одинаковые**.

### Средние пользовательские рейтинги жанров Action и Sports разные

Задача с жанрами аналогична предыдущей задаче про платформы. 
Нулевая гипотеза - средние пользовательские рейтинги игр жанров Action и Sports равны, альтернативная - средние пользовательские рейтинги игр жанров Action и Sports разные. Применим метод scipy.stats.ttest_ind(), так как генеральные совокупности не зависят друг от друга, а выборочные средние нормально распределены. Используем дополнительный параметр equal_var = False, потому что у нас недостаточно большие и неодинаковые по размеру выборки - есть основания полагать, что разброс будет сильно отличаться.

```python
action_hyp = data_new_hypothesis.query('genre == "Action"') # сделаем срез по жанру
```

```python
sports_hyp = data_new_hypothesis.query('genre == "Sports"') # сделаем срез по второму жанру
```

```python
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
results = st.ttest_ind(
    action_hyp['user_score'], 
    sports_hyp['user_score'], equal_var = False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Таким образом, полученные данные **противоречат** гипотезе о том, что средние пользовательские рейтинги игр в жанрах Action и Sports **одинаковые**.
